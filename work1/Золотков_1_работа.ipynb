# **Логическое "И"**

```python
from random import randint
```

```python
speed = 0.3
w = [randint(0,100)/100, randint(0,100)/100]
print("Стартовые веса: ",w)

x_train = [[1, 1], [1, 0], [0, 1], [0, 0]]
y_train = [1, 0, 0, 0]
```

```python
def predict(x_train):
  summator = x_train[0] * w[0] + x_train[1] * w[1]
  act = 1 if summator >= 1 else 0
  return act
```

```python
def train(x_train, y_train, w):
  for i in range(len(x_train)):
    error = y_train[i] - predict([x_train[i][0], x_train[i][1]])
    if error != 0:
      w[0] = w[0] + error * x_train[i][0] * speed
      print(w)
      w[1] = w[1] + error * x_train[i][1] * speed
      print(w)
```

```python
def epoch(count, func_train, x_train, y_train, w):
  print(f"Параметры весов:\n")
  for i in range(count):
    func_train(x_train, y_train, w)
```

```python
epoch(5, train, x_train, y_train, w)
print("\nКонечные веса: ",w)
```

```python
print(predict([1,1]))
print(predict([1,0]))
print(predict([0,1]))
print(predict([0,0]))
```

```python
from random import randint
speed = 0.3
w = [1, -1]

x_train = [[1, 1], [1, 0], [0, 1], [0, 0]]
y_train = [0, 1, 1, 0]
```

```python
def predict(x_train):
  summator = x_train[0] * w[0] + x_train[1] * w[1]
  act = abs(summator)
  act = 1 if act >= 0.5 else 0
  return act
```

```python
epoch(5, train, x_train, y_train, w)
```

```python
print(predict([1,1]))
print(predict([1,0]))
print(predict([0,1]))
print(predict([0,0]))
```
